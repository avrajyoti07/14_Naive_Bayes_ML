```python
import pandas as pd
df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis = 'columns',inplace = True)
df.head()
```

```python
# Single bracket [ ] vs Double bracket [[ ]] in Pandas & Sklearn
# df.drop() uses single brackets because it just expects a plain list of column name strings — df.drop(['col1', 'col2'], axis='columns').
# df['col'] with single brackets returns a 1D Series, which is fine for viewing data but breaks sklearn methods.
# df[['col']] with double brackets returns a 2D DataFrame, which is what sklearn's fit(), transform(), and fit_transform() always require.
# sklearn fit/transform always uses double brackets — scaler.fit_transform(df[['Age', 'Income']]) — because sklearn expects 2D input with rows as samples and columns as features.
```

```python
target = df.Survived
inputs = df.drop('Survived',axis = 'columns')
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
inputs.drop('Sex',axis = 'columns',inplace = True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(inputs,target,test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
